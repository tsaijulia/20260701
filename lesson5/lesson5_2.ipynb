```python
#錯誤的範例
score = int(input("請輸入分數"))
if score >= 90:
    grade = "優"
else:
    if score >= 80:
        grade = "甲"
    else:
        if score >= 70:
            grade = "乙"
        else:
            if score >= 60:
                grade = "丙"
            else:
                grade = "丁"
print(grade)            
```

```python
# 多項選擇
# 正確的範例
score = int(input("請輸入分數"))
if score >= 90:
    grade = "優"
elif score >= 80:
     grade = "甲"
elif score >= 70:
     grade = "乙"
elif score >= 60:
     grade = "丙"
else:
        grade = "丁"
print(grade)            
```

```python
# 在多項選擇中，條件的判定是由上而下的依序進行
age = int(input("請輸入年紀"))
if age > 18:
    print("少年")
elif age > 12:
    print("青少年")
else:
    print("青年")
```

## 異常處理 (try - except) 的資料驗證

```python
try :
    money = int(input("請輸入金額:"))
    print("輸入的金額為:",money)
except ValueError:
    print("輸入格式錯誤，請輸入正數數字!")
except:
    print("不知名的錯誤")
```

```python
# 國文100分同時數學100分，將金1000元
# 2科只有一個100分，將金500元
# 如果都沒有100分，將金0元

bonus = 0
chinese = 100
math = 100
if chinese == 100:
    if math == 100:
        bonus = 1000
    else:
        bonus = 500
else:
    if math == 100:
        bonus = 500
print(bonus) 
```

```python
# 國文100分同時數學100分，將金1000元
# 2科只有一個100分，將金500元
# 如果都沒有100分，將金0元

bonus = 0
chinese = 100
math = 100
if chinese == 100:
    if math == 100:
        bonus = 1000
    else:
        bonus = 500
elif math == 100:
        bonus = 500
print(bonus)
```

```python
# 國文100分同時數學100分，將金1000元
# 2科只有一個100分，將金500元
# 如果都沒有100分，將金0元

chinese = 100
math = 100
if chinese == 100 and math == 100:
    bonus = 1000
elif chinese == 100 or math == 100:
    bonus = 500
else:
    bonus = 0

print(bonus)
```
