## while迴圈

```python
# while1.py
a = 6
while a > 0:
    print(a)
    a -= 1 
print("離開後 a 為", a)
```

## 計算 2-10 所有的偶數的總和：while有()

```python
total = 0
i = 2
while(i <= 10):
    total += i
    print(f"第{i//2}次迴圈的i={i},總合為{total}")
    i += 2
    
```

```python
total = 0
i = 2
while i <= 10:
    total += i
    print(f"第{i//2}次迴圈的i={i},總合為{total}")
    i += 2
```

## range(結束值)
起始值(start)：預設為 0
結束值(stop)：不包含這個數字
間隔值(step)：預設為 1

```python
for x in range(5):
    print(x)
```

## range(開始, 結束)
起始值(start)：預設為 0
結束值(stop)：不包含這個數字
間隔值(step)：預設為 1

```python
for x in range(1,5):
    print(x)
```

## range(起始值, 結束值, 間隔值)
起始值(start)：預設為 0
結束值(stop)：不包含這個數字
間隔值(step)：預設為 1

```python
for x in range(1,10,2):
    print(x)
```

## 1到100的奇數與偶數加總

```python
even_total = 0
odd_total = 0
for i in range(1,101):
    if i % 2 == 0:
        even_total += i
    else:
        odd_total += i
print(f"1到100的偶數加總為{even_total}")
print(f"1到100的奇數加總為{odd_total}")
```

## 1~50中，3的倍數總和

```python
multiple_of_threeTotal = 0
notMultiple_of_threeTotal = 0

for i in range(1, 51):
    # 請完成:判斷 i 是否為 3 的倍數
    if i % 3 == 0:
        # 請完成:將 i 加到 3 的倍數總和
        multiple_of_threeTotal += i
    else:
        # 請完成:將 i 加到不是 3 的倍數總和
        notMultiple_of_threeTotal += i

print(f"1到50中，3的倍數總和為 {multiple_of_threeTotal}")
print(f"1到50中，不是3的倍數總和為 {notMultiple_of_threeTotal}")
```
