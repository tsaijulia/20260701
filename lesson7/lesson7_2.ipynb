```python
# dict 的 typehint 
country_code:dict[str,str] = {"tw" : "Taiwan", "jp" : "Japen" , "hk": "Hong Kong", "kr" : "Kroea" }
for code in country_code:
    print(f"key:{code}")
    print(f"value:{country_code[code]}")
    print("================")
```

## while 迴圈

```python
# 輸入英文字母 (按 q 離開)
while True:
    stuff = input("請輸入小寫的英文字母[按q會離開]:")
    if stuff == 'q':
        break
    print(stuff.capitalize())
print("程式結束")
```

```python
# 累加正偶數
# 設定一個程式讓使用者輸入數值，只有正偶數值會被加總，若輸入奇數則跳過，輸入負值則程式結束。

total:int = 0
num:int = 0
while True:
    input_value = int(input(f"請輸入第{num + 1}個數值:"))
    if input_value < 0:
        break
    num += 1 

    if input_value % 2 == 1:
        continue

    total += input_value
print(f"輸入的次數是{num},所有輸入的正偶數加總是:{total}")

    
```

## 巢狀迴圈

```python
for i in range(1,6):
    for j in range(i):
        print(j)
    print()
```

## 巢狀迴圈

```python
for i in range(1,6):
    for j in range(i):
        print(j, end='')
    print()
```

## 巢狀迴圈

```python
for i in range(1,6):
    for j in range(i):
        print('#', end='')
    print()
```

## 巢狀迴圈

```python
for i in range(1,6):
    for _ in range(i):
        print('#', end='')
    print()
```
